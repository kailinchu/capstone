```python
import plotly.express as px
import pandas as pd
```

```python
atmospheric_pressure = 760 
freq_hz = 60

def plot_pressure(csv_file, convert_with_atm = False):
    df = pd.read_csv(csv_file)
    df.columns = ['time (s)', 'pressure (mmHg)']

    # Adjust values
    if convert_with_atm:
        df['pressure (mmHg)'] = df['pressure (mmHg)'] - atmospheric_pressure
    
    df['time (s)'] = df['time (s)'] / freq_hz

    fig = px.line(df, x="time (s)", y="pressure (mmHg)", title='Pressure in mmHg', )
    fig.show()
    return df
```

```python
import plotly.graph_objects as go


def plot_multi_pressure(files, start_at_increase = False):
    dfs = []
    for f in files:
        df = pd.read_csv(f)
        df.columns = ['time (s)', 'pressure (mmHg)']

        # Adjust values
        df['time (s)'] = df['time (s)'] / freq_hz

        if start_at_increase:
            df_gt_0 = df[df['pressure (mmHg)'] > 0]['time (s)']
            if len(df_gt_0) > 0:
                time_offset = df_gt_0.iloc[0]
                df['time (s)'] = df['time (s)'] - time_offset
                df = df[df['pressure (mmHg)'] > 0]

        dfs.append(df)


    fig = go.Figure()

    for df, f in zip(dfs, files):
        fig.add_trace(go.Scatter(x=df['time (s)'], y=df['pressure (mmHg)'], mode='lines', name=f))

    fig.update_layout(title="Pressure Measured in Blood Pressure Cuff")
        
    fig.show()
```

```python
plot_multi_pressure(['data/0_PWM.csv', 'data/100_PWM.csv', 'data/192_PWM.csv'])
```

```python
plot_multi_pressure(['data/0_PWM.csv', 'data/100_PWM.csv', 'data/192_PWM.csv','data/stepwise_dorian.csv', 'data/stepwise_jon.csv'], start_at_increase=True)
```
